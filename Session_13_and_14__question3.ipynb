```python
import matplotlib.pyplot as pl
import numpy as np

x = np.linspace(1, 100)

def f(x):
    return np.sin(x) + np.random.normal(scale=0.1, size=len(x))

pl.plot(x, f(x))
```

```python
random_out = f(x)
```

```python
range = np.abs(np.max(random_out)-np.min(random_out))
Threshoold10 = (range/10)+np.min(random_out)
Threshoold20 = (range/5)+np.min(random_out)
i = 0
for x in random_out:
  if x <Threshoold20:
    i+=1
print(i)
print(Threshoold20)
print(range)
print(np.min(random_out))

```

```python
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=3, n_init=10)
gm.fit(random_out.reshape(-1,1))
```

```python
densities = gm.score_samples(random_out.reshape(-1,1))
density_threshold = np.percentile(densities, 4)
anomalies = random_out.reshape(-1,1)[densities < density_threshold]
```
